# Travel Destination Recommendation System

Scrapping booking.com

The project goal was to recommend user with a list of top activities along with the city and its location details. For this, a popular online website booking.com was choosen. Scrapping the website included several steps - getting the cities within North America region to scrapping each individual activities in detail. First, the initial page was scrapped for obtaining each of the cities with their urls. The urls were then visited to extract all the activities with their short details, also with the urls of each activity. Further, each of those individual activities were visited for scrapping the detailed description.

```python
def scrape_single_attraction(city_name, soup):
    name = soup.find('h3', class_='css-jv2qn6')
    name = name.text if name is not None else ""
    description = soup.find('div', class_='css-6k49yo')
    description = description.text if description is not None else ""
    duration = soup.find('div', class_='a53cbfa6de css-j786b1')
    duration = duration.text if duration is not None else ""
    rating = soup.find('span', class_='a53cbfa6de css-35ezg3')
    rating = rating.text if rating is not None else ""
    reviews = soup.find('span', class_='a53cbfa6de')
    reviews = reviews.text if reviews is not None else ""
    price = soup.find('div', class_='e1eebb6a1e css-13pzcpe')
    price = price.text if price is not None else ""
    link = soup.find('a', class_='css-i6rjpg')
    link = link['href'] if link is not None else ""
    image = soup.find('img', class_='css-17k46x')
    image = image["src"] if image is not None else ""
    
    attraction = {
        'city': city_name,
        'name': name,
        'description': description,
        'duration': duration,
        'rating': rating,
        'reviews': reviews,
        'price': price,
        'link': link,
        'image': image,
    }
    return attraction
```

```python
def scrape_attractions_city(city_name, city_url):
    driver.get(city_url)
    wait = WebDriverWait(driver, 20)

    while True:
        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(5)
            see_more = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div.css-1f31mt9 button')))
            driver.execute_script("arguments[0].scrollIntoView();", see_more)
            driver.execute_script("arguments[0].click();", see_more)
            time.sleep(5) 
            continue
        except TimeoutException:
            print('No more button to show')
            break
        except Exception as e:
            print(e)
            break
        
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    cards = soup.find_all('div', class_='b817090550 b736e9e3f4')
    city_attractions = []
    for card in cards:
        attraction = scrape_single_attraction(city_name, card)
        city_attractions.append(attraction)
    
    return city_attractions
```

# Stopword removal

```python
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

def tokenize_sentences(text):
    return sent_tokenize(text)

# Apply sentence tokenization to 'text_column'
df['description'] = df['description'].apply(tokenize_sentences)

# Explode the lists of sentences into separate rows, while keeping other columns intact
df = df.explode('description').reset_index(drop=True)
```

# Preprocessing

Lemmatization

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))
```

Splitting of text into tokens

```python
def preprocess_text(text):
    tokens = text.lower().split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)
```

```python
df['processed_description'] = df['description'].apply(preprocess_text)
```

```python
df['processed_description'].value_counts()
```

# Feature Extraction

For encoding of the features into respective number vectors, TfIDF was used. TfIDF is one of the most widely used and popular feature extraction techniques. TfIDF not only considers the frequency of a word, but also finds it rarety. 

```python
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
```

# Model Evaluation

Function to recommend activity based on the probability. Since, the predictions are to be made on the user input, matching those with the probability maps to the processed description. 

```python
def recommend_activity(input_description, top_n,best_classifier,test_activity):
    processed_input = preprocess_text(input_description)
    input_tfidf = tfidf_vectorizer.transform([processed_input])
    input_features = input_tfidf.toarray()
    
    best_probs = best_classifier.predict_proba(input_features)[0]
    top_indices = np.argsort(best_probs)[::-1][:top_n]
    top_activities = best_classifier.classes_[top_indices]
    top_probs = best_probs[top_indices]
    

    recommendations = []
    for activity, probability in zip(top_activities, top_probs):
        # activity = df[df['activity'] == activity]['activity']
        # recommendations.append(activity)

        activity = df.loc[df['activity'] == activity, 'activity']
        recommendations.append(activity)

        
    for item in recommendations:
        if item.iloc[0] == test_activity:
           return True
    return False
```

The models were evaluated based on the count ratio of total correct recommendations to that of the total number of recommendations, over a subset of the dataset. This checks if the model is aligning with the original dataset or not.

```python
import random

def evaluate_model(classifier, dataframe):
    print("Performance Evaluation:")
    recommendation_count =5
    true_recommendation = 0
    false_recommendation = 0
    count =300
    random_rows = dataframe.sample(n=count, random_state=42)
    for index,row in random_rows.iterrows():
        num_words = 20
        input_description = row["processed_description"].split()
        input_description = ' '.join(input_description[0:num_words])

        res = recommend_activity(input_description,recommendation_count,classifier,row["activity"])
        # print(count,res)
        if res:
            true_recommendation+=1
        else:
            false_recommendation+=1
    accuracy =true_recommendation/count       
    print("Acccuracy: ", round(accuracy,2))
    return(accuracy)
```

# Recommender

Function to recommend the activity, city and providing the location.

```python
def recommend_city(input_description, top_n=5):
    processed_input = preprocess_text(input_description)
    input_tfidf = tfidf_vectorizer.transform([processed_input])
    input_features = input_tfidf.toarray()
    
    best_probs = best_classifier.predict_proba(input_features)[0]
    top_indices = np.argsort(best_probs)[::-1][:top_n]
    top_cities = best_classifier.classes_[top_indices]
    top_probs = best_probs[top_indices]
    
    
    recommendations = []
    for activity, probability in zip(top_cities, top_probs):
        city = df[df['activity'] == activity]['city']
        price =df[df["activity"] == activity]["price"]
        try:
            location = df[df['activity'] == activity]['location']
        except KeyError:
            location = "Location not available"
            
        try:
            link = df[df['activity'] == activity]['link']
        except KeyError:
            link = "url not available"
        recommendations.append((activity, probability, city, location,link,price))
        
    return recommendations


```

# User Interface

The project idea was to provide a simple input option to the user and output the recommendations where user would be able to see the desired attractions along with the cities and their location with respective urls. For the purpose of the user interface- Tkinter was used. Tkinter is a simple and useful Python GUI library, having intuitive widgets and event handling for small applications.

```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import webbrowser

class RecommendationApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Travel Recommendation App")
        self.master.geometry("800x600")
        self.master.resizable(False, False)

        # Putting the widgets

        self.label = ttk.Label(master, text="Enter your description:")
        self.label.grid(row=0, column=0, padx=10, pady=10, sticky="w")

        self.input_entry = ttk.Entry(master, width=130)
        self.input_entry.grid(row=1, column=0, padx=10, pady=5)

        self.recommend_button = ttk.Button(master, text="Recommend", command=self.show_recommendations)
        self.recommend_button.grid(row=2, column=0, padx=10, pady=5)

        self.output_text = tk.Text(master, height=25, width=80)
        self.output_text.grid(row=3, column=0, padx=10, pady=10, sticky="ew")
        self.output_text.tag_configure("link", foreground="blue", underline=True)
        self.output_text.tag_bind("link", "<Button-1>", self.on_link_click)

    # Function to show recommendation

    def show_recommendations(self):
        input_description = self.input_entry.get()
        if input_description:
            try:
                recommendations = recommend_city(input_description)
                self.output_text.delete(1.0, tk.END)
                self.output_text.insert(tk.END, "Travel Recommendations\n")
                self.output_text.insert(tk.END, "+" + "-"*88 + "+\n")
                self.output_text.insert(tk.END, "| {:<20} | {:<20} | {:<30} |\n".format("City", "Price", "Activity"))
                self.output_text.insert(tk.END, "+" + "-"*88 + "+\n")
                count = 0
                for activity, probability, city, location, link, price in recommendations:
                    count+=1
                    city_text = "{:<2}| {:<20} | {:<20} | {:<30} |\n".format(count,city.iloc[0], price.iloc[0], activity)
                    self.output_text.insert(tk.END, city_text)
                    self.output_text.insert(tk.END, "+" + "-"*88 + "+\n")

                    # Adding clickable link
                    self.output_text.insert(tk.END, "| {:<73} |\n".format("Link: " + link.iloc[0]), "link")
                    self.output_text.insert(tk.END, "+" + "-"*88 + "+\n\n\n")

            except Exception as e:
                messagebox.showerror("Error", str(e))
        else:
            messagebox.showwarning("Warning", "Please enter a description.")

    # redirecting code if the link is clicked

    def on_link_click(self, event):
        index = self.output_text.index("@%s,%s" % (event.x, event.y))
        tag = self.output_text.tag_names(index)
        if "link" in tag:
            link_start = self.output_text.search("Link: ", index, backwards=True, regexp=True)
            if link_start:
                link_start = link_start + "+6c"
                link_end = self.output_text.search("\n", link_start)
                if link_end:
                    link = self.output_text.get(link_start, link_end)
                    webbrowser.open_new(link)

def main():
    root = tk.Tk()
    app = RecommendationApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
